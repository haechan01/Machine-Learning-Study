## Q1: What is Naive Bayes Model?

### Summary

The **Naïve Bayes model** is a family of simple yet effective probabilistic classifiers based on applying Bayes' Theorem with a strong assumption of independence among features given the class label. It calculates the probability of each class for a given instance and predicts the class with the highest probability. This model is widely used in tasks like text classification, spam filtering, and sentiment analysis due to its efficiency and effectiveness.

---

### Explanation

The Naïve Bayes model utilizes **Bayes' Theorem**, which describes the probability of an event based on prior knowledge of related conditions:

$$
P(C \mid X) = \frac{P(X \mid C) \cdot P(C)}{P(X)}
$$

Where:

- $P(C \mid X)$: Posterior probability of class $C$ given feature vector $X$.
- $P(X \mid C)$: Likelihood of features $X$ given class $C$.
- $P(C)$: Prior probability of class $C$.
- $P(X)$: Marginal probability of feature vector $X$.

In classification, $P(X)$ is constant across classes and can be ignored for comparison. The model computes the posterior probability for each class and selects the class with the highest value.

The "naïve" aspect comes from assuming that all features $x_i$ in $X = (x_1, x_2, \dots, x_n)$ are **conditionally independent** given the class label $C$. This simplifies the likelihood computation:

$$
P(X \mid C) = \prod_{i=1}^{n} P(x_i \mid C)
$$

This assumption reduces the complexity of calculating joint probabilities, making the model computationally efficient, especially with high-dimensional data.

**Types of Naïve Bayes Classifiers:**

1. **Gaussian Naïve Bayes**: Assumes continuous features follow a Gaussian distribution.
2. **Multinomial Naïve Bayes**: Used for discrete data; common in document classification where features represent word counts.
3. **Bernoulli Naïve Bayes**: Suitable for binary features indicating the presence or absence of a feature.

---

### Example

**Spam Email Classification**

Imagine developing a spam filter that classifies emails as "Spam" or "Not Spam" based on the words they contain.

1. **Feature Extraction**:
   - Convert each email into a feature vector where each feature represents the presence (1) or absence (0) of specific words from a vocabulary.

2. **Training Phase**:
   - **Calculate Priors**:
     - $P(\text{Spam})$: Proportion of spam emails in the training set.
     - $P(\text{Not Spam})$: Proportion of non-spam emails.
   - **Calculate Likelihoods**:
     - For each word $w_i$ in the vocabulary:
       - $P(w_i = 1 \mid \text{Spam})$: Probability that word $w_i$ appears in spam emails.
       - $P(w_i = 1 \mid \text{Not Spam})$: Probability that word $w_i$ appears in non-spam emails.

3. **Classification Phase**:
   - For a new email, extract the feature vector $X$.
   - Compute posterior probabilities for each class:

     $$
     P(\text{Spam} \mid X) \propto P(\text{Spam}) \cdot \prod_{i=1}^{n} P(x_i \mid \text{Spam})
     $$

     $$
     P(\text{Not Spam} \mid X) \propto P(\text{Not Spam}) \cdot \prod_{i=1}^{n} P(x_i \mid \text{Not Spam})
     $$

   - Assign the email to the class with the higher posterior probability.

**Handling Zero Probabilities:**

To avoid zero probabilities when a word in the new email wasn't seen in the training data, **Laplace smoothing** is used:

$$
P(w_i \mid C) = \frac{\text{Count}(w_i, C) + 1}{\text{Total Words in } C + V}
$$

Where:

- $\text{Count}(w_i, C)$: Number of times word $w_i$ appears in class $C$.
- $V$: Size of the vocabulary.

## Q2: What makes a naive Bayes "naive"?

### Summary

Naïve Bayes is called "naïve" because it makes a strong assumption that all features are **conditionally independent** of each other given the class label. This means the presence or absence of a particular feature is assumed to be unrelated to any other feature, given the class. In reality, this assumption is often unrealistic, hence the term "naïve."

---

### Explanation

The Naïve Bayes classifier applies **Bayes' Theorem** with the simplifying assumption of feature independence:

$$
P(C \mid X) = \frac{P(X \mid C) \cdot P(C)}{P(X)}
$$

Where:

- $P(C \mid X)$: Posterior probability of class $C$ given features $X$.
- $P(X \mid C)$: Likelihood of features $X$ given class $C$.
- $P(C)$: Prior probability of class $C$.
- $P(X)$: Marginal probability of features $X$.

Under the **naïve assumption**, the likelihood simplifies to:

$$
P(X \mid C) = \prod_{i=1}^{n} P(x_i \mid C)
$$

Each feature $x_i$ is considered independent of the others given the class label $C$. This greatly simplifies computation but ignores any potential dependencies between features, making the model "naïve."

## Q3: What's a latent variable?
### Summary

A **latent variable** is a variable that is not directly observed but is inferred from other variables that are observed (measured). In machine learning and statistics, latent variables represent hidden or unobserved factors that influence the observed data.

---

### Explanation

Latent variables are used to model complex systems where certain variables are not directly measurable. They help capture underlying structures or patterns in the data by accounting for hidden influences.

In probabilistic models, latent variables allow us to express the joint distribution of observed variables \( X \) and latent variables \( Z \) as:

$$
P(X, Z) = P(X \mid Z) \, P(Z)
$$

Here:

- \( X \): Observed variables (data).
- \( Z \): Latent variables (hidden factors).
- \( P(Z) \): Prior distribution of the latent variables.
- \( P(X \mid Z) \): Likelihood of observing \( X \) given \( Z \).

By integrating out the latent variables, we obtain the marginal distribution of the observed data:

$$
P(X) = \int P(X, Z) \, dZ
$$

Latent variables are fundamental in models like:

- **Hidden Markov Models (HMMs)**
- **Gaussian Mixture Models (GMMs)**
- **Latent Dirichlet Allocation (LDA)**
- **Variational Autoencoders (VAEs)**

They enable these models to capture hidden structures and make probabilistic inferences about the data.

## Q4: What's a case where the naïve assumption breaks down?
### Summary
# Summary

The naïve assumption in Naïve Bayes breaks down when features are **highly correlated** or **dependent** even after conditioning on the class label. In such cases, the assumption of conditional independence is violated, leading to inaccurate probability estimates and poor model performance.

---

# Explanation

When features are dependent given the class label \( C \), the simplification:

$$
P(X \mid C) = \prod_{i=1}^{n} P(x_i \mid C)
$$

does not hold true. Instead, the joint probability \( P(X \mid C) \) cannot be accurately represented as a simple product of individual probabilities.

**Situations where the assumption breaks down:**

- **Multicollinearity**: Features provide overlapping or redundant information.
- **Sequential Data**: In text or time series data, the order of features matters.
- **Interaction Effects**: The presence of one feature changes the effect of another.

Violating the independence assumption can cause the classifier to underestimate or overestimate the likelihoods, leading to incorrect predictions.

## Q5: Explain the difference between a joint probability and marginal probability.

### Summary

**Joint probability** refers to the probability of two or more events occurring simultaneously, denoted as \( P(A, B) \). **Marginal probability** is the probability of a single event occurring regardless of the outcomes of other events, obtained by summing or integrating over the joint probabilities, denoted as \( P(A) \).

---

### Explanation

- **Joint Probability (\( P(A, B) \))**:
  - Represents the probability of events \( A \) and \( B \) happening at the same time.
  - For discrete variables:
    $$
    P(A = a, B = b)
    $$
  - For continuous variables:
    $$
    P(A = a, B = b) = f_{A,B}(a, b)
    $$
    where \( f_{A,B}(a, b) \) is the joint probability density function.

- **Marginal Probability (\( P(A) \))**:
  - The probability of event \( A \) occurring, irrespective of event \( B \).
  - Obtained by summing (discrete case) or integrating (continuous case) the joint probability over all possible values of \( B \):
    - Discrete case:
      $$
      P(A = a) = \sum_{b} P(A = a, B = b)
      $$
    - Continuous case:
      $$
      P(A = a) = \int_{-\infty}^{\infty} f_{A,B}(a, b) \, db
      $$

**Key Difference**:

- **Joint Probability** considers the probability of multiple events occurring together.
- **Marginal Probability** focuses on a single event, summing over the probabilities of all possible outcomes of other events.

## Q6: When would the joint probability not equal the product of the marginals? 

### Summary

The joint probability \( P(A, B) \) does not equal the product of the marginal probabilities \( P(A)P(B) \) when events \( A \) and \( B \) are **dependent**. If the occurrence of one event affects the probability of the other, their joint probability differs from the product of their marginals.

---

### Explanation

- **Independent Events**: When events \( A \) and \( B \) are independent:
  $$
  P(A, B) = P(A) \times P(B)
  $$
  
- **Dependent Events**: When events \( A \) and \( B \) are dependent:
  $$
  P(A, B) \neq P(A) \times P(B)
  $$
  Instead, the joint probability is calculated using conditional probability:
  $$
  P(A, B) = P(A \mid B) \times P(B)
  $$
  or
  $$
  P(A, B) = P(B \mid A) \times P(A)
  $$
  
- **Reason**: Dependence means the occurrence of one event influences the likelihood of the other. Therefore, the joint probability must account for this relationship using conditional probabilities.

## Q7: What is the connection between how partitioned vector spaces (Trees) relate to Bayesian probability, and how joint distributions can be compactly calculated with matrix multiplications (linear algebra)

### Summary

Partitioned vector spaces represented as trees in graphical models (like Bayesian networks) capture conditional dependencies between variables in Bayesian probability. These trees allow for the joint distribution to be factorized based on the structure, simplifying computations. Joint distributions can then be compactly calculated using matrix multiplications in linear algebra, leveraging these factorizations for efficient computation.

---

### Explanation

**Partitioned Vector Spaces and Trees:**

- **Graphical Models**: Trees in graphical models represent variables as nodes and conditional dependencies as edges.

- **Bayesian Networks**: A type of graphical model where the joint probability distribution is factorized according to the network structure.

- **Factorization**: For variables \( X = (X_1, X_2, \dots, X_n) \), the joint distribution is:

  $$
  P(X) = \prod_{i} P(X_i \mid \text{Parents}(X_i))
  $$

  - \( \text{Parents}(X_i) \): The set of parent nodes of \( X_i \) in the tree.
  - This factorization reflects the partitioning of the vector space according to conditional dependencies.

**Matrix Multiplications and Joint Distributions:**

- **Linear Algebra**: Matrices can represent probability distributions and transformations between them.

- **Compact Calculations**: Joint distributions over multiple variables can be computed using matrix multiplications, especially in models like Hidden Markov Models (HMMs).

  - **State Transition Matrix \( A \)**: Represents probabilities of transitioning between states.

    $$
    A_{ij} = P(S_{t} = j \mid S_{t-1} = i)
    $$

  - **Emission Matrix \( B \)**: Represents probabilities of observations given states.

    $$
    B_{jk} = P(O_{t} = k \mid S_{t} = j)
    $$

- **Efficiency**: Using matrices allows leveraging efficient linear algebra operations to compute joint probabilities over sequences.

# Core Question (3): SMS Classification

Pretend it's the early 2010's where people still text each other over SMS. You're kind of tired of all the spam texts from your cell provider offering discounts, random things you signed up for and the occasional creepy stranger. We can build a classifier for whether it's "ham" or "spam". We'll be using this dataset, consisting of 615 "spam" and 4957 "ham" texts.

```python
# load in dataset
import pandas as pd
from sklearn.feature_extraction import text
df = pd.read_csv("https://raw.githubusercontent.com/milindsoorya/Spam-Classifier-in-python/main/dataset/spam.csv", encoding='latin-1')
labels, texts = df['v1'], df['v2']
vectorizer = text.TfidfVectorizer()
vec_texts = vectorizer.fit_transform(texts)

# Uncomment and print one at a time
# print(texts[:3])
# Hint: vec_texts is a sparse matrix; the coordinates are only shown for its "nonzero" entries
#print(vec_texts[:3], vec_texts[:3].shape)
#print(vectorizer.vocabulary_)
```

## Question 2 of 6
Above, we loaded in some text and converted it into vectors. What's this "TfidfVectorizer" thing doing? Look it up, and summarize it in about 100 words or so.

TfidfVectorizer converts text into numerical feature vectors using TF-IDF weighting, which stands for Term Frequency-Inverse Document Frequency. This method calculates how often a word appears in a document and scales it down based on how common the word is across all documents. This way, important words in each document are emphasized while common words are given less weight.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics, naive_bayes

# We train a model
model = naive_bayes.MultinomialNB()
model.fit(vec_texts, labels)
pred_labels = model.predict(vec_texts)

confusion_mat = metrics.confusion_matrix(labels, pred_labels)
sns.heatmap(confusion_mat, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

How is Naive Bayes being "learned" here? Give a numerical answer for how many total parameters you expect in this model, and how you arrived at that figure.

Naïve Bayes learns by estimating probabilities for each feature independently within each class. Since the TF-IDF conversion results in 8,672-dimensional vectors and we have two classes (e.g., "spam" and "not spam"), the model learns parameters for each feature-class combination.

If we're using Gaussian Naïve Bayes (suitable for continuous features like TF-IDF scores), we need to estimate the mean and variance of each feature for each class. 

Calculation of the total number of parameters:

Number of features: 8,672
Number of classes: 2
Parameters per feature per class: 2 (mean and variance)
Total parameters = Number of features × Number of classes × Parameters per feature per class
Total parameters = 8,672 × 2 × 2 = 34,688

TThus, we expect the model to learn 34,688 parameters in total.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(
    "https://raw.githubusercontent.com/milindsoorya/Spam-Classifier-in-python/main/dataset/spam.csv",
    encoding='latin-1'
)
labels = df['v1']
texts = df['v2']

# Convert text data into TF-IDF vectors
vectorizer = TfidfVectorizer()
vec_texts = vectorizer.fit_transform(texts)

# Train a Multinomial Naive Bayes model
model = MultinomialNB()
model.fit(vec_texts, labels)

# Get the feature names (tokens)
feature_names = vectorizer.get_feature_names_out()

# Get the log probabilities of features given each class
# model.feature_log_prob_ has shape (n_classes, n_features)
# Classes are ordered alphabetically by default ('ham', 'spam')
# We need to find the index of the 'spam' class
spam_class_index = np.where(model.classes_ == 'spam')[0][0]

# Extract log probabilities for the 'spam' class
spam_feature_log_prob = model.feature_log_prob_[spam_class_index]

# Convert log probabilities to probabilities
spam_feature_prob = np.exp(spam_feature_log_prob)

# Create a DataFrame for easy handling
token_probs = pd.DataFrame({
    'token': feature_names,
    'probability': spam_feature_prob
})

# Sort tokens by probability in descending order
top_tokens = token_probs.sort_values(by='probability', ascending=False).head(30)

# Plot the top 30 tokens
plt.figure(figsize=(12, 6))
plt.bar(top_tokens['token'], top_tokens['probability'])
plt.xticks(rotation=90)
plt.title('Top 30 Tokens with Highest P(X | "spam")')
plt.ylabel('P(X | "spam")')
plt.xlabel('Tokens')
plt.tight_layout()
plt.show()
```

```python
# Sort tokens by probability in ascending order
bottom_tokens = token_probs.sort_values(by='probability', ascending=True).head(30)

# Plot the bottom 30 tokens
plt.figure(figsize=(12, 6))
plt.bar(bottom_tokens['token'], bottom_tokens['probability'])
plt.xticks(rotation=90)
plt.title('Bottom 30 Tokens with Lowest P(X | "spam")')
plt.ylabel('P(X | "spam")')
plt.xlabel('Tokens')
plt.tight_layout()
plt.show()
```


In the first plot, the tokens with the highest $P(X \mid \text{"spam"})$ values are words commonly associated with spam content, such as **"call," "free,"  "now," "claim,"** and **"prize."** These words often relate to promotions or offers that spam messages frequently contain to call customers into action.

Conversely, the tokens with the lowest $P(X \mid \text{"spam"})$ values are words more commonly found in legitimate "ham" messages and hard to be found in "spam messages". These words are more often used in daily settings, but not as often in spam messages. Also, I see some proper nouns and typos which would not usually occur in spam messages. Some of the examples are **"nanny," "ned," "nattil," "nap,"** and **"nevr."** Such words are associated with everyday communication or accidental typos.

**Factors Influencing Token Likelihood:**

- **Frequency in Spam vs. Ham Messages:** Tokens that appear frequently in spam messages but rarely in ham messages have higher $P(X \mid \text{"spam"})$ values and contribute more to a message being classified as spam.
- **Content Type:** Words related to promotions, financial incentives, urgency, or actions are more indicative of spam.
- **Common Language in Legitimate Communication:** Tokens that are common in regular, non-spam communications contribute to a message being classified as ham. These also include proper nouns and words with spelling error.
